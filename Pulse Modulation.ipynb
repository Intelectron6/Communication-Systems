```python
# Pulse Modulation

import numpy as np
import matplotlib.pyplot as plt
```

```python
#Message signal and rectangular pulse
t = np.linspace(0,100,1000)
m = 5 + 5*np.cos(2*3.14*30*t)
p = np.concatenate(10*(np.zeros(80),np.ones(20)))
```

```python
plt.plot(t, m)
plt.show()
plt.plot(t, p)
plt.show()
```

```python
#Pulse Amplitude Modulation
s = m*p
plt.plot(t, s)
plt.show()
```

```python
#Pulse Code Modulation

#Sampling
m_disc = []
for i,a in enumerate(m):
    if i%50 == 0:
        m_disc.append(a)
k = np.linspace(0,100,20)
m_disc = np.array(m_disc)
plt.bar(k, m_disc)
plt.show()
```

```python
#Quantization
n = 3
L = 2**n
delta = (max(m_disc)-min(m_disc))/L
delta = round(delta*100)/100
```

```python
levels = [delta*i for i in range(L)]
```

```python
m_quant = []
for i,a in enumerate(m_disc):
    diff = []
    for j,b in enumerate(levels):
        diff.append(abs(b-a))
    i = diff.index(min(diff))
    m_quant.append(levels[i])
m_quant = np.array(m_quant)
```

```python
plt.bar(k, m_quant)
plt.show()
```

```python
#Encoding
ones = [5 for i in range(100)]
zeros = [0 for i in range(100)]
```

```python
pcm = []
for a in m_quant:
    if a == 8.75:
        pcm.extend(ones+ones+ones)
    if a == 7.5:
        pcm.extend(ones+ones+zeros)
    if a == 6.25:
        pcm.extend(ones+zeros+ones)
    if a == 5.0:
        pcm.extend(ones+zeros+zeros)
    if a == 3.75:
        pcm.extend(zeros+zeros+ones)
    if a == 2.5:
        pcm.extend(zeros+ones+zeros)
    if a == 1.25:
        pcm.extend(zeros+zeros+ones)
    if a == 0.0:
        pcm.extend(zeros+zeros+zeros)
```

```python
plt.plot(pcm)
```
