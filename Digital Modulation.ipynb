```python
# Bandpass transmission of Digital Signals

import numpy as np
import matplotlib.pyplot as plt
```

```python
p = np.concatenate(2*(np.zeros(250),np.ones(250)))
t = np.linspace(0,100,1000)
c = 5*np.cos(2*3.14*1000*t)
```

```python
plt.plot(t, c)
plt.show()
plt.plot(t, p)
plt.show()
```

```python
#Amplitude Shift Keying
ask = p*c
plt.plot(t, ask)
plt.show()
```

```python
#Frequency Shift Keying
c2 = 5*np.cos(2*3.14*400*t)
plt.plot(t, c2)
plt.show()

pi = []
for i,a in enumerate(p):
    if a == 1:
        pi.append(0)
    else:
        pi.append(1)

f1 = p*c
f2 = pi*c2
fsk = f1 + f2
plt.plot(t, fsk)
plt.show()
```

```python
#Phase Shift Keying
p_polar = 0.5 - p
fsk = p_polar*c2
plt.plot(t, fsk)
plt.show()
```
