```python
#Time domain reperesentation of FM and PM

import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,100,10000)
c = 3*np.cos(2*3.14*1000*t)
m = 5*np.cos(2*3.14*100*t)
```

```python
# Narrow Band FM
snb = 3*np.cos(2*3.14*1000*t + 0.75*np.sin(2*3.14*100*t))
plt.plot(t,snb)
plt.show()
```

```python
# Wide Band FM
swb = 3*np.cos(2*3.14*1000*t + 7.5*np.sin(2*3.14*100*t))
plt.plot(t,swb)
plt.show()
```

```python
# PM
sp = 3*np.cos(2*3.14*1000*t + 2.5*m)
plt.plot(t,sp)
plt.show()
```
