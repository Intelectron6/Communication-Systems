```python
#Time domain reperesentation of various AM schemes

import numpy as np
import matplotlib.pyplot as plt
#from scipy.fftpack import fft, ifft
```

```python
#Carrier and message signals

t = np.linspace(0,100,10000)
c = 3*np.cos(2*3.14*1000*t)
m = 5*np.cos(2*3.14*100*t)
```

```python
plt.plot(t,c)
plt.show()
plt.plot(t,m)
plt.show()
```

```python
#Standard Amplitude Modulation
sam = 5*(1+(3/5)*m)*c
```

```python
plt.plot(t,sam)
plt.show()
```

```python
#Double Side Band Suppressed Carrier
dsbsc = m*c
```

```python
plt.plot(t,dsbsc)
plt.show()
```

```python
#Hilbert transforms of carrier and message
ch = 3*np.sin(2*3.14*1000*t)
mh = 5*np.sin(2*3.14*100*t)
```

```python
#Single Side Band Suppressed Carrier
ssb_l = (m*c + mh*ch)/2
ssb_u = (m*c - mh*ch)/2
```

```python
plt.plot(t,ssb_l)
plt.show()
plt.plot(t,ssb_u)
plt.show()
```

```python
#Demodulation of each of the modulated signals
mr1 = sam*c
```

```python
plt.plot(t,mr1)
plt.show()
```

```python
mr2 = dsbsc*c
```

```python
plt.plot(t,mr2)
plt.show()
```

```python
mr3 = ssb_u*c
mr4 = ssb_l*c
```

```python
plt.plot(t,mr3)
plt.show()
plt.plot(t,mr4)
plt.show()
```
